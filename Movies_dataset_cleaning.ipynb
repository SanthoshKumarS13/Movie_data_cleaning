 IF WE are handeling the data have \nAction, Horror, Thriller "\n" this kind of special characters means we can go with regex=False so Using regex=False can help avoid unintended behavior when replacing strings that might contain special characters. For instance, if your data contains characters like '.' or '*', these have special meanings in regular expressions. By setting regex=False, you ensure that these characters are treated as literals.
